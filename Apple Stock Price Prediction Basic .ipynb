```python
import numpy as np
import matplotlib.pyplot as plt

# Generating artificial dataset
np.random.seed(42)  # For reproducibility

# Creating a linear relationship with some random noise
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Visualizing the dataset
plt.scatter(X, y, color='blue', label='Data Points')
plt.title('Artificial Dataset for Linear Regression')
plt.xlabel('Sales')
plt.ylabel('Price')
plt.legend()
plt.show()

```

```python
from sklearn.linear_model import LinearRegression

# Creating a Linear Regression model
model = LinearRegression()

# Fitting the model to the data
model.fit(X, y)

# Making predictions using the trained model
y_pred = model.predict(X)

# Visualizing the linear regression line
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, y_pred, color='red', label='Linear Regression Line')
plt.title('Linear Regression on Artificial Dataset')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Fetching historical data for Apple stock
apple_data = yf.download('AAPL', start='2020-01-01', end='2024-01-01')

# Extracting closing prices as the dependent variable (y)
y = apple_data['Close'].values.reshape(-1, 1)

# Creating an array of sequential numbers as the independent variable (X)
X = np.arange(len(y)).reshape(-1, 1)

# Creating a Linear Regression model
model = LinearRegression()

# Fitting the model to the data
model.fit(X, y)

# Making predictions using the trained model
y_pred = model.predict(X)

# Visualizing the linear regression line
plt.scatter(X, y, color='blue', label='Actual Stock Prices')
plt.plot(X, y_pred, color='red', label='Linear Regression Line')
plt.title('Linear Regression on Apple Stock Prices')
plt.xlabel('Days Since Start')
plt.ylabel('Closing Price')
plt.legend()
plt.show()

```
